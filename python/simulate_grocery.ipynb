```python
import pandas as pd
import numpy as np

orig_data = pd.read_csv('data/food-footprints.csv')
orig_data
```

```python
orig_dict = orig_data[["Entity", "GHG emissions per kilogram (Poore & Nemecek, 2018)"]].to_dict()
```

```python
item_pairs = list(zip(orig_dict["Entity"].values(), orig_dict["GHG emissions per kilogram (Poore & Nemecek, 2018)"].values()))
```

```python
sim_list = []

for i in range(len(item_pairs)):
    k = np.random.randint(3, 10)
    for v in range(k):
        co2_mean = item_pairs[i][1]
        co2 = np.random.normal(co2_mean, co2_mean/5, 1)[0]
        
        if co2 <= 0:
            co2 = co2_mean
        
        weight = np.random.normal(co2_mean * 30, co2_mean * 30/5, 1)[0]
        price = np.random.normal(weight * 5, weight/2, 1)[0]
        calories = np.random.normal(weight * 6, weight/3, 1)[0]
        health_score = np.random.normal(price * 4, weight/3, 1)[0]

        sim_list.append([item_pairs[i][0], str(item_pairs[i][0]) + "-" + str(v), round(weight, 2) , round(price, 2), round(calories, 2), round(health_score, 2), round(co2, 2) ])
```

```python
sim_list
```

```python
sim_list_df = pd.DataFrame(sim_list)
sim_list_df.columns = ["category", "article", "weight", "price", "calories", "health_score", "co2"]
```

```python
np.min(sim_list_df[['co2']].values.flatten())
```

```python
np.mean(sim_list_df[['calories']].values.flatten())
```

```python
sim_list_df.to_csv("data/sim_food.csv")
```
